```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_title('title')
axes.set_xlabel('x')
axes.set_ylabel('y')
```

```python
fig2 = plt.figure()
axes1 = fig2.add_axes([0,0,1,1])
axes2 = fig2.add_axes([0.2,0.5,0.2,0.2])
```

```python
axes1.plot(x,y)
axes1.set_xlabel('x')
axes1.set_ylabel('y')

axes2.plot(x,y)
axes2.set_xlabel('x')
axes2.set_ylabel('y')

fig2
```

```python
fig3 = plt.figure()
ax1 = fig3.add_axes([0,0,1,1])
ax2 = fig3.add_axes([0.2,0.5,.4,.4])
```

```python
ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.plot(x,y)
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
fig3
```

```python
fig4,axes = plt.subplots(nrows=1,ncols=2)
```

```python
axes[0].plot(x,y,color='blue',lw=3,ls=':')
axes[1].plot(x,z,color='yellow',lw=2,ls='-')
fig4
```

```python
fig4,axes = plt.subplots(nrows=1,ncols=2,figsize=(4,2))
axes[0].plot(x,y,color='blue',lw=3,ls=':')
axes[1].plot(x,z,color='yellow',lw=2,ls='-')
```
